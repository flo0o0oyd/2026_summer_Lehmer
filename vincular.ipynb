# Reverse and Complement of a Permutation

```python
def rev_comp(perm):
    n = len(perm)
    temp = perm[::-1]
    ret_val = [n+1-x for x in temp]
    return Permutation(ret_val)

def rev(perm):
    return Permutation(perm[::-1])

def comp(perm):
    n = len(perm)
    ret_val = [n+1-x for x in perm]
    return Permutation(ret_val)
```

# Avoiding vincular patterns of length 3

```python
def avoids_321(perm):
    n = len(perm)
    for i in range(n-2):
        if perm[i] > perm[i+1] and perm[i+1] > perm[i+2]:
                return False
    return True

def avoids_3_21(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+1,n-1):
            if perm[i] > perm[j] and perm[j] > perm[j+1]:
                return False
    return True

def avoids_32_1(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+2,n):
            if perm[i] > perm[i+1] and perm[i+1] > perm[j]:
                return False
    return True

def avoids_1_23(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+1,n-1):
            if perm[i] < perm[j] and perm[j] < perm[j+1]:
                return False
    return True

def avoids_12_3(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+2,n):
            if perm[i] < perm[i+1] and perm[i+1] < perm[j]:
                return False
    return True 

def avoids_21_3(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+2,n):
            if perm[i] > perm[i+1] and perm[i] < perm[j]:
                return False
    return True

def avoids_23_1(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+2,n):
            if perm[i] < perm[i+1] and perm[i] > perm[j]:
                return False
    return True

def avoids_1_32(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+1,n-1):
            if perm[i] < perm[j+1] and perm[j] > perm[j+1]:
                return False
    return True    

def avoids_3_12(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+1,n-1):
            if perm[i] > perm[j+1] and perm[j] < perm[j+1]:
                return False
    return True 
 
    
```

# maximum weight 3_21 avoiding permutations

```python
for n in range(1,8):
    
    # get 3_21 avoiding permutations
    my_perm = []
    for p in Permutations(n):
        if avoids_321(p):
            my_perm.append(p)
    
    max_inv = 0
    max_list = []
    for p in my_perm:
        my_inv = p.number_of_inversions()
        if my_inv > max_inv:
            max_inv = my_inv
            max_list = [ p,]
        elif my_inv == max_inv:
            max_list.append(p)
    
    print('===', n, max_inv, len(max_list))
    for m in max_list: 
        print('\t', m.to_lehmer_code())
```

## Formulas for count and weight

```python
def tri(k):
    return binomial(k+1,2)

def rho(n):
    return floor((-3 + sqrt(17+8*n))/2)

def lam(n):
    return n + 2 - tri(rho(n)+1)
```

```python
def count(n):
    return binomial(rho(n),lam(n)) + binomial(rho(n)+1, lam(n)-1)

def weight(n):
    temp = [ tri(i) for  i in range(1, rho(n)+1)]
    
    return sum(temp) + tri(n-rho(n))
    
```

```python
def is_valid(code):
    size = len(code)
    for j in range(size-1):
        if code[j] > code[j+1]:
            for i in range(j):
                if code[j]+j - code[i]-i <= 0:
                    return False
    return True
```
