# Reverse and Complement of a Permutation

```python
def rev_comp(perm):
    n = len(perm)
    temp = perm[::-1]
    ret_val = [n+1-x for x in temp]
    return Permutation(ret_val)

def rev(perm):
    return Permutation(perm[::-1])

def comp(perm):
    ret_val = [n+1-x for x in perm]
    return Permutation(ret_val)
```

# Avoiding vincular patterns of length 3

```python
def avoids_3_21(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+1,n-1):
            if perm[i] > perm[j] and perm[j] > perm[j+1]:
                return False
    return True

def avoids_32_1(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+2,n):
            if perm[i] > perm[i+1] and perm[i+1] > perm[j]:
                return False
    return True

def avoids_1_23(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+1,n-1):
            if perm[i] < perm[j] and perm[j] < perm[j+1]:
                return False
    return True

def avoids_12_3(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+2,n):
            if perm[i] < perm[i+1] and perm[i+1] < perm[j]:
                return False
    return True 

def avoids_21_3(perm):
    n = len(perm)
    for i in range(n-2):
        for j in range(i+2,n):
            if perm[i] > perm[i+1] and perm[i] < perm[j]:
                return False
    return True    

 
    
```

# maximal jump-float sequences

```python
def filter_maximal(sequences):
    maximal = []
    for a in sequences:
        dominated = False
        for b in sequences:
            if a == b:
                continue
            # b dominates a if b[i] >= a[i] for all i, strictly greater somewhere
            if all(bj >= aj for aj, bj in zip(a, b)) and any(bj > aj for aj, bj in zip(a, b)):
                dominated = True
                break
        if not dominated:
            maximal.append(a)
    return maximal
```

# maximum weight 3_21 avoiding permutations

```python
for n in range(1,8):
    
    # get 3_21 avoiding permutations
    my_perm = []
    for p in Permutations(n):
        if avoids_3_21(p):
            my_perm.append(p)
    
    max_inv = 0
    max_list = []
    for p in my_perm:
        my_inv = p.number_of_inversions()
        if my_inv > max_inv:
            max_inv = my_inv
            max_list = [ p,]
        elif my_inv == max_inv:
            max_list.append(p)
    
    print('===', n, max_inv, len(max_list))
    for m in max_list: 
        print('\t', m.to_lehmer_code())
```

## Formulas for count and weight

```python
def tri(k):
    return binomial(k+1,2)

def rho(n):
    return floor((-3 + sqrt(17+8*n))/2)

def lam(n):
    return n + 2 - tri(rho(n)+1)
```

```python
def count(n):
    return binomial(rho(n),lam(n)) + binomial(rho(n)+1, lam(n)-1)

def weight(n):
    temp = [ tri(i) for  i in range(1, rho(n)+1)]
    
    return sum(temp) + tri(n-rho(n))
    
```
