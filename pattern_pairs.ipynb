```python
def code_comparator(a,b):
    for x,y in zip(a,b):
        if x > y:
            return False
    return True
```

```python
# returns set of maximal Lehmer codes
def get_maximal_set(perm_list):
    maximal_set = set()

    for p in perm_list:
        g = tuple(p.to_lehmer_code())
        is_covered = False
        for m in maximal_set:
            if code_comparator(g,m):
                is_covered = True
                break
        if not is_covered:
            remove_set = set()
            for m in maximal_set:
                if code_comparator(m,g):
                    remove_set.add(m)
            for r in remove_set:
                maximal_set.remove(r)
            maximal_set.add(g)
    return maximal_set

def get_maximum_weight_set(lehmer_list):
    max_weight = 0
    max_set = set()
    
    for g in lehmer_list:
        if sum(g) > max_weight:
            max_weight = sum(g)
            max_set = set()
            max_set.add(g)
        elif sum(g) == max_weight:
            max_set.add(g)
    return max_set

def enumerate_stats(perm_list):
        maximal_set = get_maximal_set(perm_list)
        max_wt_set = get_maximum_weight_set(maximal_set)
        print('n=', n, 'num perm=', len(perm_list), 'num maximal=', len(maximal_set), 'num max weight', len(max_wt_set))
```

# Poset Construction

```python
for n in range(5,6):
    good_list = []
    for p in Permutations(n, avoiding=[[3,2,1],[2,1,3]]):
        good_list.append(p.to_lehmer_code())
    print('== n', n, 'num',len(good_list))
    for g in good_list:
        print('\t', g)
    good_list = [tuple(x) for x in good_list]
    P = Poset((good_list, code_comparator))
    me = P.maximal_elements()
    for m in me:
        print(m)
    print(P.is_lattice())
    print(P.is_join_semilattice())
    print(P.is_meet_semilattice())
    print(P.is_graded())
    print(P.is_ranked())
    print(P.is_series_parallel()) 
    P.show()
```
